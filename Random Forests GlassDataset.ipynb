```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

# The Dataset

```python
data = pd.read_csv('E:\\DataScience\\Phyton\\ml\\DecisionTrees\\RandomForests\\glass_data.csv')
data
```

```python
data = data.drop(['Id'], axis = 1)
data
```

```python
inputs = data.iloc[:,:-1]
target = data.iloc[:,-1]
```

```python
inputs.head()
```

# Splitting Data

```python
train_inputs, test_inputs, train_target, test_target = train_test_split(inputs, target, test_size = 0.2, random_state = 152)
```

# Creating Random Forest Model

```python
clf = RandomForestClassifier()
clf = clf.fit(train_inputs,train_target)
```

```python
clf.predict([train_inputs.iloc[99]])
```

# Accuracy

```python
test_target_pred = clf.predict(test_inputs)
```

```python
print(classification_report(test_target, test_target_pred))
```
