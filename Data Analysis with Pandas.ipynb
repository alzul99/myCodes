***Data Manipulation***

```python
list_MOS_time =[]
```

```python
for i in range(0,len(sales_data['TimeZone'])):
    list_MOS_time.append(pd.to_datetime(sales_data.iloc[i].loc['MOS'].tz_localize(sales_data.iloc[i].loc['TimeZone'])))
list_MOS_time
```

```python
len(list_MOS_time)
```

```python
list_MOS_time[1].strftime('%z')
```

```python
list_MOS_time[1].utcoffset().total_seconds()/3600
```

```python
sales_data['Offset_UTC'] = [list_MOS_time[i].utcoffset().total_seconds()/3600 for i in range(len(list_MOS_time))]
```

```python
sales_data
```

```python
sales_data['MOS_UTC']=[sales_data['MOS'].iloc[1].tz_localize(sales_data['TimeZone'].iloc[1]).tz_convert('UTC') for i in range(len(sales_data['MOS']))]
```

```python
sales_data.head(5)
```

```python
columns_reorder = ['RetailerCountry','MOS_UTC','MOS','TimeZone','Offset_UTC','RetailerType','Product','Sales Revenue ($)']
```

```python
sales_data = sales_data[columns_reorder]
```

***Data Visualisation***

```python
import matplotlib.pyplot as plt
```

```python
sales_data_viz = sales_data.reset_index()
```

```python
sales_data_viz.head()
```

```python
sales_data_viz['Offset_UTC'].plot(figsize=(10, 6))
```

```python
plt.scatter(sales_data_viz['Offset_UTC'], sales_data_viz['RetailerType'])
```

```python
sales_data_viz.groupby(['TimeZone']).sum()
```

```python
Total_Revenue_by_TimeZone = sales_data_viz.groupby(['TimeZone']).sum()
Total_Revenue_by_TimeZone
```

```python
Total_Revenue_by_TimeZone.iloc[:,1]
```

```python
plt.bar(Total_Revenue_by_TimeZone.index,
       height = Total_Revenue_by_TimeZone.iloc[:,1])
```

```python
plt.bar(Total_Revenue_by_TimeZone.index,
       height = Total_Revenue_by_TimeZone.iloc[:,1],
       color ='red')
plt.xticks(rotation = 45, fontsize =12)
plt.yticks(fontsize = 12)
plt.title('Total Sales Revenue by Time Zone', weight = 'bold', fontsize = 15)
plt.ylabel('Sales Revenue ($)', fontsize =12)
plt.show()
```

```python
Averages_by_timezone = sales_data_viz.groupby(['TimeZone']).mean().sort_values('Sales Revenue ($)', ascending = False)
Averages_by_timezone
```

```python
plt.figure(figsize=(10,8))
plt.bar(Averages_by_timezone.index,
       height = Averages_by_timezone.iloc[:,1],
       color =['green','red','blue','brown','purple','cyan'])
plt.xticks(rotation = 45, fontsize =12)
plt.yticks(fontsize = 12)
plt.title('Average Sales Revenue by Time Zone', weight = 'bold', fontsize = 15)
plt.ylabel('Sales Revenue ($)', fontsize =12)
plt.show()
```
